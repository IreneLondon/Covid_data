```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
cvc_df = pd.read_csv('covid19vaccinesbycounty.csv')
cvc_df
```

```python
cvc_df['county'].value_counts()
                              
```

```python
cvc_df['administered_date'].value_counts()
```

```python
cvc_df['Administered Month'] = pd.DatetimeIndex(cvc_df['administered_date']).month
```

```python
cvc_df['Administered Year'] = pd.DatetimeIndex(cvc_df['administered_date']).year
```

```python
plt.hist(cvc_df['Administered Year'],bins = 30)
plt.show()
```

```python
plt.hist(cvc_df['Administered Month'],bins = 30)
plt.show()
```

```python
cvc_df['Administered Year'].value_counts()
```

```python
plt.hist(cvc_df['total_doses'],bins = 30)
plt.show()
```

```python
plt.boxplot(cvc_df['total_doses'])
plt.show()
```

```python
plt.figure(figsize = (15,10))
cvc_df.groupby('Administered Year').sum()['total_doses'].plot(kind = 'pie',autopct = '%1.1f%%')
```

```python
plt.figure(figsize = (15,10))
cvc_df.groupby('Administered Year').sum()['pfizer_doses'].plot(kind = 'pie',autopct = '%1.1f%%')
```

```python
plt.figure(figsize = (15,10))
cvc_df.groupby('Administered Year').sum()['jj_doses'].plot(kind = 'pie',autopct = '%1.1f%%')
```

```python
plt.figure(figsize = (15,10))
cvc_df.groupby('Administered Year').sum()['total_doses'].plot(kind = 'pie',autopct = '%1.1f%%')
```

```python
plt.figure(figsize = (15,10))
cvc_df.groupby('Administered Year').sum()['moderna_doses'].plot(kind = 'pie',autopct = '%1.1f%%')
```

```python
plt.figure(figsize = (15,10))
cvc_df.groupby('Administered Month').sum()['Administered Year'].plot(kind = 'pie',autopct = '%1.1f%%')
```
